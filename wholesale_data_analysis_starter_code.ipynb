## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df_part_two = df
df_part_two.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
#if unit_weight > 50, then charge $7; if unit_weight < 50, then charge $10

# Calculate total weight for each row
total_weight = df_part_two['unit_weight'] * df_part_two['qty']

# Apply the shipping price based on the condition
df_part_two['shipping_price'] = total_weight.apply(lambda x: 7 if x > 50 else 10)
df_part_two['shipping_price'] = (df_part_two['shipping_price'] * df_part_two['unit_weight'])
new_df = ['first', 'last', 'job', 'phone', 'email', 'client_id', 'order_id',
       'order_date', 'order_week', 'order_year', 'item_id', 'category',
       'subcategory', 'unit_price', 'unit_cost', 'unit_weight', 'qty',
       'line_number', 'shipping_price', 'subtotal']
df_part_two.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

df_part_two['shipping_price'] = df_part_two['shipping_price'].astype(float)

df_part_two['total_price'] = (df_part_two['subtotal'] + df_part_two['shipping_price']) * 1.0925
df_part_two['total_price'] = df_part_two['total_price'].round(2)

df_part_two.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and shipping price (assume the shipping cost is exactly what is charged to the client).

# Create a column for the cost of each line. Use unit cost, qty, and ...
# In this case "line" is synonymous with "record" or "row". Better wording altogether:
# Create a column that calculates total cost (including shipping) for each row.

df_part_two['line_cost'] = df_part_two['unit_cost'] * df_part_two['qty'] + df_part_two['shipping_price']
df_part_two.head()
```

```python
# Create a column for the profit of each line using line cost and line price

df_part_two['line_profit'] = df_part_two['total_price'] - df_part_two['line_cost']
df_part_two.head()
```
